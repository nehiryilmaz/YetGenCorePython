# Problem 1 
Kullanıcıdan aldığı ürün bilgisini (ad, fiyat) urunler.txt dosyasına kayıt eden fonksiyonu yazınız.

````python
def urun_kaydet() :
    urunAdi = input("Ürünün adı :")
    urunFiyati = int(input("Ürünün fiyatı :"))
    with open("urunlar.txt","a",encoding="utf-8") as file :
        file.write(f"{urunAdi} : {urunFiyati}")
    print("Ürün kaydedildi.")
urun_kaydet()
# 1.Problemin Çözümünü Buraya Yazınız
````

````python
def kelime_degistir(dosyaadi, eski_kelime, yeni_kelime) :
    with open("urunlar.txt","r",encoding="utf-8") as file :
        dosya = file.read()
    yeni_dosya = dosya.replace(eski_kelime,yeni_kelime)
    with open("urunlar.txt","w",encoding="utf-8") as file :
        file.write(yeni_dosya)
    print("Kelime değiştirildi.") 
dosyaadi = "urunlar.txt"
eski_kelime = input("Değiştirilecek olan kelimeyi giriniz :")
yeni_kelime = input("Yeni kelimeyi giriniz :")
kelime_degistir(dosyaadi, eski_kelime, yeni_kelime)
# 2.Problemin Çözümünü Buraya Yazınız
````
